```python
import pandas as pd
from finvizfinance.quote import finvizfinance
from finvizfinance.screener.overview import Overview
from finvizfinance.screener.financial import Financial
from finvizfinance.screener.valuation import Valuation
```

```python
foverview = Overview()
filter_1 = {'Country':'USA', 'Market Cap.': '+Mid (over $2bln)'} #'Index':'S&P 500', 
foverview.set_filter(filters_dict=filter_1)
df = foverview.screener_view()
df.count()
```

```python
#keep 1000 largest cap
df=df.nlargest(1000, ['Market Cap'])
#drop rows for real estate
df = df[df.Sector != 'Real Estate']
#drop useless columns 
df=df.drop(columns=['Industry','Country','Market Cap','P/E','Price','Change','Volume'])
```

```python
#add yield column
#first download fundamentals with same filter
ffinancial = Financial()
filter_1 = {'Country':'USA', 'Market Cap.': '+Mid (over $2bln)'} #'Index':'S&P 500', 
ffinancial.set_filter(filters_dict=filter_1)
df2 = ffinancial.screener_view()
#Then, join
df2.count()
```

```python
#Then, join
df3 = pd.merge(df, df2, how='left', on=['Ticker'])
df3.count()
```

```python
#keep 200 largest yields
df4=df3.nlargest(200, ['Dividend'])
#keep only useful columns
df4 = df4[['Ticker', 'Company', 'Sector','Dividend', 'ROA']]
df4.count()
```

```python
#keep the top 4 ROA by sector
#sorting first
df4 = df4.sort_values(['Sector', 'ROA'], ascending=False)
df5=df4.groupby('Sector').head(4)
df5.head(5)
```

```python
#keeping the top20 ROA
df5 = df5.sort_values(['ROA'], ascending=False)
df6=df5.head(20).sort_values(['Sector'], ascending=False).reset_index(drop=True)
df6
```
